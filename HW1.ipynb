```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as skt
import numpy as np
```

```python
# reading in csv file for data
ny_data = pd.read_csv("nyt7.csv")
ny_data = pd.DataFrame(ny_data)
```

## Case Study: New York Times Ad Impressions
1. Create a new variable, `age_group`, that categorizes users into the following age groups: < 18, 18-24, 25-34, 35-44, 45-54, 55-64, and 65+. 

```python
# can use np.max(ny_data['Age']) to see the highest age is 112
ny_data["age_groups"] = pd.cut(ny_data["Age"],
                               bins = [-1,18,24,34,44,54,64,112],
                               labels=['<18','18-24','25-34','35-44','45-54','55-64','65+'])


# ensure age_groups is a categorical variable
ny_data["age_groups"] = pd.Categorical(ny_data["age_groups"],
                                       categories = ['<18','18-24','25-34','35-44','45-54','55-64','65+'],
                                       ordered=True)
```

2. Plot the distributions of impressions and "click-through rate" for all 6 age categories. (Note: Click-through rate is defined as the number of clicks divided by the number of impressions; it represents the proportion of ads that generated clicks.)

```python
# plotting distribution of impressions
sns.histplot(x = 'Impressions', hue = 'age_groups', element = 'step', data = ny_data)

plt.xticks(np.arange(np.min(ny_data["Impressions"]), np.max(ny_data["Impressions"]) + 1))

# Add labels and title
plt.xlabel('Impressions')
plt.ylabel('Density')
plt.title('Density Plot with Histogram of Impressions across Age Groups')

# Show the plot
plt.show()
```
