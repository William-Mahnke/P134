```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# read in starwars data from R
starwars = pd.read_csv("~/Desktop/P134/starwars.csv")
```

```python
starwars
```

```python
# 1. How many rows are in starwars? How many columns?
rows, cols = starwars.shape
print(f"Rows: {rows}\nColumns: {cols}")

# 2. What does `vehicles` variable describe?
# describes the vehicles used by a character in the show - was stored as a list in R which
# is why the column in all NaNs
```

```python
starwars_smaller = starwars.loc[:,['name','height','species','gender']]
starwars_smaller = starwars_smaller[starwars_smaller['height'] >= 0]
starwars_smaller
```

```python
# 1. What is the purpose of `starwars_smaller['height']` is?
starwars_smaller['height'] >= 0
# assigns boolean to each observation which allows for flitering particular
# observations
starwars_smaller['height'].describe()
starwars_smaller.groupby('gender')['height'].describe()
```

```python
# histogram of character heights
plt.hist(starwars_smaller['height'])
plt.title("Distribution of Character Heights")
plt.xlabel("Height in cm")
plt.show()
```

```python
# histogram and density plot split by gender
sns.histplot(data = starwars_smaller, x = 'height', hue = 'gender',
             multiple = 'stack', bins = 15, kde = True)
plt.title("Distribution of Height by Genders")
plt.legend(title = 'Gender', labels = ['Female','Male'])
plt.show()
```
